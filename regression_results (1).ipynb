This file contains the code to generate regression results from the experiment, the code generates 
linear plot to demonstrate the relationship between the ground truth and the prediction.

Note: you need to run the experiment and generate the gt.npy and preds.npy before running this notebook

I have ran the experiments in 6 different networks and plotted them below. To run new experiment, all you
need to do is changing the path and run the same code again

```python
"""
This notebook contains the code to generate regression results from the experiment, the code generates
linear plot to demonstrate the relationship between the ground truth and the prediction
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

base_cwd = "/home/azureuser/cloudfiles/code/Users/tingyan.deng"
exp_name = "res18_withbos"
exp_dir  = base_cwd + "/gsv-ml-buildings-new-2-inception/exps/" + exp_name + "/"

gt    = np.load(exp_dir+"gt.npy")
preds = np.load(exp_dir+"preds.npy")
```

```python
#resnet34
plt.scatter(gt, preds)
plt.xlim(18, 35)
plt.ylim(18, 35)
plt.plot([0, 60], [0, 60], color="red")
m, b = np.polyfit(gt, preds, 1)
plt.plot(gt, m*gt+b, color="orange")

plt.show()
#res34
```
